```r
#install libraries
install.packages("tidyverse")
install.packages("ggpubr")
```

```r
#import the data file using url

data <- read.csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
```

```r
summary(data)
```

```r
x <- data$Hours
y <- data$Scores

# Plot 

plot(x, y, main = "Hours vs Scores",
     xlab = "Hours", ylab = "Scores",
     pch = 19, frame = FALSE)

# Add regression line
plot(x, y, main = "Hours vs Scores",
     xlab = "Hours", ylab = "Scores",
     pch = 19, frame = FALSE)
abline(lm(y ~ x, data = data), col = "blue")
```

```r
cor(data$Scores, data$Hours)
```

```r
model <- lm(Scores ~ Hours, data = data)
model
```

```r
summary(model)
```

```r
predict(model, data.frame(Hours = 9.25))
```
