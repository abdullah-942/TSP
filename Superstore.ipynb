### Data Exploration

```python
# Import the packages

import pandas as pd
import numpy as np
```

```python
#Load the file

store = pd.read_csv("C:/Users/aishaque/Downloads/SampleSuperstore.csv")
```

```python
#There is only 1 country in the country column. So drop taht column.
#Also, drop the Postal Code because its of no use in EDA

store = store.drop(['Country', 'Postal Code'], axis=1)
```

### Data Cleaning

```python
#check if there is any missing value

store.isnull().sum()
```

```python
#Check for row duplicacy

store.duplicated().sum()
```

```python
#Remove the duplicated rows

store.drop_duplicates(inplace = True)
```

```python
store
```

```python
##Let us group sales, quantity, and profit under one header.

store_profit = store[['Sales','Quantity','Profit']].groupby(['Profit'],as_index=True).sum().sort_values(by='Quantity',ascending=True)
```

```python
# Importing visualization libraries

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Visualize the 'Segment' column from the dataset.

sns.set_style('whitegrid')
sns.countplot(x='Segment',data = store, palette='rainbow')
```

```python
#Visualize the 'Category' column from the dataset.

sns.set_style('whitegrid')
sns.countplot(x='Category',data = store, palette='rainbow')
```

```python
#Visualize the 'SubCategory' column from the dataset.

sns.set_style('whitegrid')
sns.countplot(x='Sub-Category',data = store, palette='rainbow')
```

```python
#Visualize the 'Region' column from the dataset.

sns.set_style('whitegrid')
sns.countplot(x='Region',data = store, palette='rainbow')
```

```python
#Visualize the 'Ship Mode' column from the dataset.

sns.set_style('whitegrid')
sns.countplot(x='Ship Mode',data = store, palette='rainbow')
```

### Bivariate Analysis

```python
#Visualize the 'Region' column from the Ship mode.

region = sns.FacetGrid(store, col='Ship Mode', palette='rainbow')
region.map(plt.hist, 'Region')
region.set_ylabels('Number')

```

```python
#Visualize the Category and Ship Mode.

category = sns.FacetGrid(store, col='Ship Mode', palette='rainbow')
category.map(plt.hist, 'Category')
category.set_ylabels('Number')

```

```python
#Visualize the SubCategory and Segment.

subcategory = sns.FacetGrid(store, col='Segment', palette='rainbow')
subcategory.map(plt.hist, 'Sub-Category')
subcategory.set_ylabels('Number')

```

```python
#Visualize the 'Category' column from the Shipmode.

a = sns.FacetGrid(store, col='Category', row='Sub-Category')
a.map(plt.hist, 'Segment')
a.set_ylabels('Number')
a.add_legend()
```

```python
z = sns.FacetGrid(store, row='Category', col='Sub-Category')
z.map(sns.barplot, 'Profit', 'Segment', alpha=.8, ci=None)
z.add_legend()
```
